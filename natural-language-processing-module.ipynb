Esse código em Python realiza uma análise de repetições de palavras, frases e expressões em um DataFrame usando a biblioteca Pandas. 

Importação de bibliotecas:

pandas: é uma biblioteca usada para manipulação e análise de dados.
matplotlib.pyplot: é uma biblioteca usada para plotagem de gráficos.
collections.Counter: é uma classe que conta a ocorrência de elementos em uma sequência.
Função encontrar_palavras_repetidas:

Recebe um DataFrame (df) como entrada.
Extrai todas as palavras das mensagens do DataFrame, transformando-as em letras minúsculas.
Conta a ocorrência de cada palavra usando a classe Counter.
Retorna uma lista das palavras repetidas (que ocorrem mais de uma vez).
Função encontrar_frases_repetidas:

Recebe um DataFrame (df) como entrada.
Extrai todas as frases das mensagens do DataFrame.
Conta a ocorrência de cada frase usando a classe Counter.
Retorna um dicionário das frases repetidas (que ocorrem mais de uma vez), juntamente com a quantidade de ocorrências.
Função encontrar_expressoes_repetidas:

Recebe um DataFrame (df) e uma lista de expressões (expressoes) como entrada.
Transforma todas as expressões e mensagens em letras minúsculas.
Conta quantas vezes cada expressão ocorre nas mensagens.
Retorna uma lista das expressões repetidas (que ocorrem mais de uma vez).
Função plotar_grafico_percentual:

Recebe um dicionário de dados (dados) e um título (titulo) como entrada.
Extrai os valores e as chaves do dicionário.
Calcula a porcentagem de cada valor em relação ao total.
Plota um gráfico de pizza usando a biblioteca matplotlib.pyplot.
Exemplo de uso:

Chama as funções anteriores para encontrar palavras repetidas, frases repetidas e expressões repetidas.
Calcula os percentuais totais de palavras, frases e expressões.
Mostra as palavras repetidas com suas quantidades e percentuais.
Mostra as frases repetidas com suas quantidades.
Mostra as expressões repetidas.
Cria dicionários com os dados para os gráficos.
Plota os gráficos de percentual de palavras, frases e expressões repetidas.
Calcula os percentuais individuais de palavras, frases e expressões repetidas.
Mostra os percentuais individuais de palavras, frases e expressões repetidas.
O código tem como objetivo identificar palavras, frases e expressões que ocorrem mais de uma vez em um DataFrame e fornecer informações sobre sua quantidade e porcentagem. Também é feita a plotagem de gráficos para visualizar os percentuais de repetição.

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import json
import seaborn as sns
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

def remover_stopwords(texto):
    stopwords_portugues = set(stopwords.words('portuguese'))
    palavras = texto.lower().split()
    palavras_sem_stopwords = [palavra for palavra in palavras if palavra not in stopwords_portugues]
    texto_sem_stopwords = ' '.join(palavras_sem_stopwords)
    return texto_sem_stopwords

def encontrar_palavras_repetidas(df):
    palavras = " ".join(df['mensagem']).lower().split()
    palavras_sem_stopwords = [remover_stopwords(palavra) for palavra in palavras]
    contador_palavras = Counter(palavras_sem_stopwords)
    palavras_repetidas = [palavra for palavra, ocorrencias in contador_palavras.items() if ocorrencias > 1]
    return palavras_repetidas


def encontrar_frases_repetidas(df):
    frases = df['mensagem'].tolist()
    contador_frases = Counter(frases)
    total_frases = len(frases)
    frases_repetidas = {frase: [ocorrencias, (ocorrencias / total_frases) * 100]
                        for frase, ocorrencias in contador_frases.items() if ocorrencias > 1}
    return frases_repetidas

def encontrar_expressoes_repetidas(df, expressoes):
    expressoes_lower = [expressao.lower() for expressao in expressoes]
    mensagens_lower = [mensagem.lower() for mensagem in df['mensagem']]
    expressoes_repetidas = [expressao for expressao in expressoes_lower if mensagens_lower.count(expressao) > 1]
    return expressoes_repetidas

def plotar_grafico_percentual(dados, titulo):
    valores = list(dados.values())
    labels = list(dados.keys())

    total = sum(valores)
    porcentagens = [(valor / total) * 100 for valor in valores]

    # Configurar estilo do seaborn
    sns.set(style="whitegrid")

    # Plotar o gráfico de pizza
    plt.figure(figsize=(8, 6))
    plt.pie(porcentagens, labels=labels, autopct='%1.1f%%')
    plt.title(titulo)

    # Mostrar o gráfico
    plt.show()

# Exemplo de uso
#df_atualizado = pd.read_csv('dados.csv')  # Substitua pelo nome do seu arquivo CSV

palavras_repetidas = encontrar_palavras_repetidas(df_atualizado)
frases_repetidas = encontrar_frases_repetidas(df_atualizado)
expressoes_repetidas = encontrar_expressoes_repetidas(df_atualizado, mensagens_cartao_credito)

# Cálculo dos percentuais
total_palavras = len(df_atualizado['mensagem'].str.split().sum())
total_frases = len(df_atualizado['mensagem'])
total_expressoes = len(mensagens_cartao_credito)  # Certifique-se de ter definido essa variável

# Mostrar palavras repetidas com quantidade e percentual
print("Palavras Repetidas:")
contador_palavras = Counter(df_atualizado['mensagem'].str.lower().str.split().sum())
for palavra in palavras_repetidas:
    quantidade = contador_palavras[palavra]
    percentual = (quantidade / total_palavras) * 100
    print(f"- {palavra} ({quantidade} vezes, {percentual:.2f}%)")

# Mostrar frases repetidas, quantidade e percentual
print("\nFrases Repetidas:")
for frase, (ocorrencias, percentual) in frases_repetidas.items():
    print(f"- {frase} ({ocorrencias} vezes, {percentual:.2f}%)")

# Mostrar expressões repetidas
print("\nExpressões Repetidas:")
for expressao in expressoes_repetidas:
    print("-", expressao)

# Criação dos dicionários com os dados para o gráfico
dados_palavras = {
    'Palavras Repetidas': len(palavras_repetidas),
    'Palavras Únicas': total_palavras - len(palavras_repetidas)
}

dados_frases = {
    'Frases Repetidas': len(frases_repetidas),
    'Frases Únicas': total_frases - len(frases_repetidas)
}

dados_expressoes = {
    'Expressões Repetidas': len(expressoes_repetidas),
    'Expressões Únicas': total_expressoes - len(expressoes_repetidas)
}

# Cálculo dos percentuais das palavras, frases e expressões repetidas
percentual_palavras_repetidas = (len(palavras_repetidas) / total_palavras) * 100

# Cálculo do percentual das frases repetidas
total_ocorrencias_frases = sum(ocorrencias for ocorrencias, _ in frases_repetidas.values())
percentual_frases_repetidas = (total_ocorrencias_frases / total_frases) * 100

# Cálculo dos percentuais das expressões repetidas
percentual_expressoes_repetidas = (len(expressoes_repetidas) / total_expressoes) * 100

# Mostrar os percentuais das palavras, frases e expressões repetidas
print(f"\nPercentual de Palavras Repetidas: {percentual_palavras_repetidas:.2f}%")
print(f"Percentual de Frases Repetidas: {percentual_frases_repetidas:.2f}%")
print(f"Percentual de Expressões Repetidas: {percentual_expressoes_repetidas:.2f}%")

# Salvar os dados em um arquivo JSON
dados = {
    'Palavras': {
        'Palavras Repetidas': len(palavras_repetidas),
        'Palavras Únicas': total_palavras - len(palavras_repetidas)
    },
    'Frases': {
        'Frases Repetidas': len(frases_repetidas),
        'Frases Únicas': total_frases - len(frases_repetidas)
    },
    'Expressões': {
        'Expressões Repetidas': len(expressoes_repetidas),
        'Expressões Únicas': total_expressoes - len(expressoes_repetidas)
    }
}

with open('dados.json', 'w') as arquivo_json:
    json.dump(dados, arquivo_json)

print("Arquivo JSON salvo com sucesso!")

# Plotagem dos gráficos
plotar_grafico_percentual(dados['Palavras'], 'Percentual de Palavras Repetidas')
plotar_grafico_percentual(dados['Frases'], 'Percentual de Frases Repetidas')
plotar_grafico_percentual(dados['Expressões'], 'Percentual de Expressões Repetidas')


```

O código fornecido realiza a análise de frequência de n-gramas em um conjunto de mensagens. Vou explicar passo a passo o que cada parte do código faz:

Importação das bibliotecas: O código importa as bibliotecas necessárias para realizar as análises, incluindo pandas, nltk, unicodedata, string, matplotlib.pyplot e seaborn.

Função "calcular_ngrams": Essa função recebe um texto e um valor "n" como entrada e retorna uma lista de n-gramas. Primeiro, ela normaliza o texto para remover acentos e caracteres especiais. Em seguida, remove pontuações, tokeniza o texto em palavras individuais e remove as chamadas "stop words" (palavras comuns que geralmente não contribuem para a análise, como "o", "a", "de", etc.). Por fim, a função gera os n-gramas a partir das palavras restantes e os retorna em uma lista.

Dicionários de frequência de n-gramas: São criados três dicionários vazios para armazenar as contagens de ocorrências de cada tipo de n-grama: bigramas, trigramas e quadrigramas.

Loop para processar as mensagens: O código itera sobre cada mensagem presente no dataframe chamado "df_atualizado".

Cálculo dos n-gramas: Para cada mensagem, o código chama a função "calcular_ngrams" para calcular os bigramas, trigramas e quadrigramas presentes nessa mensagem.

Atualização dos dicionários de frequência: Para cada n-grama calculado, o código verifica se ele já está presente no dicionário correspondente. Se estiver, incrementa a contagem de ocorrências; caso contrário, adiciona o n-grama ao dicionário com contagem igual a 1.

Função "plotar_grafico_barras": Essa função é usada para plotar gráficos de barras com as frequências dos n-gramas. Ela recebe o dicionário de frequência, um título para o gráfico e um valor opcional "top_n" que define a quantidade de n-gramas a serem exibidos no gráfico (padrão é 10). A função converte o dicionário em um dataframe, ordena os n-gramas pela frequência em ordem decrescente e seleciona os n-gramas com as maiores frequências. Em seguida, plota um gráfico de barras usando a biblioteca seaborn.

Plotagem dos gráficos: O código chama a função "plotar_grafico_barras" três vezes, passando os dicionários de frequência de bigramas, trigramas e quadrigramas, juntamente com os títulos apropriados.

O resultado final é a exibição de três gráficos de barras, mostrando a frequência dos principais bigramas, trigramas e quadrigramas nas mensagens do dataframe. Essa análise pode ajudar a identificar padrões linguísticos relevantes nos textos analisados.

```python
import pandas as pd
import nltk
from nltk import ngrams
from nltk.corpus import stopwords
import unicodedata
import string
import matplotlib.pyplot as plt
import seaborn as sns

# Função para calcular n-grams em uma mensagem
def calcular_ngrams(texto, n):
    texto = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('utf-8')
    texto = texto.translate(str.maketrans('', '', string.punctuation))
    
    tokens = nltk.word_tokenize(texto)
    stop_words = set(stopwords.words('portuguese'))
    tokens = [token for token in tokens if token.lower() not in stop_words]
    
    ngrams_list = []
    if len(tokens) >= n:  # Check if there are enough tokens to generate n-grams
        ngrams_list = list(ngrams(tokens, n))
    
    return ngrams_list


# Dicionários para armazenar as contagens de ocorrências de cada n-grama
frequencia_bigramas = {}
frequencia_trigramas = {}
frequencia_quadrigramas = {}

# Loop para processar cada mensagem no dataframe
for mensagem in df_atualizado['mensagem']:
    # Calcular bigramas
    bigramas = calcular_ngrams(mensagem, 2)
    for bigrama in bigramas:
        if bigrama in frequencia_bigramas:
            frequencia_bigramas[bigrama] += 1
        else:
            frequencia_bigramas[bigrama] = 1
    
    # Calcular trigramas
    trigramas = calcular_ngrams(mensagem, 3)
    for trigrama in trigramas:
        if trigrama in frequencia_trigramas:
            frequencia_trigramas[trigrama] += 1
        else:
            frequencia_trigramas[trigrama] = 1
    
    # Calcular quadrigramas
    quadrigramas = calcular_ngrams(mensagem, 4)
    for quadrigrama in quadrigramas:
        if quadrigrama in frequencia_quadrigramas:
            frequencia_quadrigramas[quadrigrama] += 1
        else:
            frequencia_quadrigramas[quadrigrama] = 1

# Função para plotar gráfico de barras com quantidade limitada de n-gramas
def plotar_grafico_barras(frequencia, titulo, top_n=10):
    df_frequencia = pd.DataFrame(list(frequencia.items()), columns=['N-grama', 'Frequência'])
    df_frequencia = df_frequencia.sort_values('Frequência', ascending=False).head(top_n)

    # Converter os n-gramas em strings
    df_frequencia['N-grama'] = df_frequencia['N-grama'].apply(lambda x: ' '.join(x))

    plt.figure(figsize=(10, 6))
    sns.barplot(data=df_frequencia, x='N-grama', y='Frequência', color='skyblue')
    plt.title(titulo)
    plt.xlabel('N-grama')
    plt.ylabel('Frequência')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

# Plotar gráficos dos bigramas, trigramas e quadrigramas
plotar_grafico_barras(frequencia_bigramas, 'Frequência de Bigramas', top_n=20)
plotar_grafico_barras(frequencia_trigramas, 'Frequência de Trigramas', top_n=20)
plotar_grafico_barras(frequencia_quadrigramas, 'Frequência de Quadrigramas', top_n=20)
```

O código fornecido realiza algumas operações de processamento de texto e geração de gráficos usando a biblioteca NetworkX e Matplotlib. Vou explicar o código em português para facilitar a compreensão:

A função plotar_grafo(grafo, top_n=10) é responsável por plotar um grafo usando as bibliotecas NetworkX e Matplotlib. Ela recebe como entrada um grafo (grafo) e um parâmetro opcional top_n (padrão para 10) para especificar o número de nós principais a serem exibidos. A função plota o grafo completo e, em seguida, plota apenas os top_n nós mais importantes em um gráfico separado.

A função preprocess_text(text) é responsável por pré-processar um texto, removendo stopwords, pontuação e acentuação. Ela utiliza as bibliotecas unicodedata, string, nltk e stopwords para realizar as etapas de pré-processamento. O texto pré-processado é retornado como uma string.

O código pressupõe a existência de um DataFrame chamado df_atualizado e adiciona uma nova coluna a ele chamada 'mensagem_preprocessada'. Essa coluna contém a versão pré-processada da coluna 'mensagem' do DataFrame. A função preprocess_text é aplicada a cada elemento da coluna 'mensagem' usando o método apply.

Três dicionários (frequencia_bigramas, frequencia_trigramas, frequencia_quadrigramas) são definidos para armazenar as contagens de ocorrência de cada n-grama (bigrama, trigrama e quadrigrama).

O código percorre cada mensagem pré-processada na coluna 'mensagem_preprocessada' do DataFrame. Para cada mensagem, ele calcula os bigramas, trigramas e quadrigramas usando a função calcular_ngrams (não fornecida no trecho de código) e atualiza os dicionários correspondentes (frequencia_bigramas, frequencia_trigramas, frequencia_quadrigramas) com as contagens de ocorrência de cada n-grama.

Após o loop, o código cria três gráficos (grafo_bigramas, grafo_trigramas, grafo_quadrigramas) com base nas chaves dos dicionários de n-gramas usando a função criar_grafo_ngramas (não fornecida no trecho de código).

Por fim, o código chama a função plotar_grafo três vezes, passando cada grafo (grafo_bigramas, grafo_trigramas, grafo_quadrigramas) e um valor top_n de 10 para plotar os gráficos e os 10 nós mais importantes em cada grafo.

```python
import networkx as nx

def criar_grafo_ngramas(ngramas):
    grafo = nx.Graph()
    for ngrama in ngramas:
        tokens = str(ngrama).split()
        for i in range(len(tokens) - 1):
            source = tokens[i]
            target = tokens[i + 1]
            if grafo.has_edge(source, target):
                grafo[source][target]['weight'] += 1
            else:
                grafo.add_edge(source, target, weight=1)
    return grafo


def plotar_grafo(grafo, top_n=10):
    plt.figure(figsize=(12, 8))
    pos = nx.spring_layout(grafo, seed=42, k=0.5)
    nx.draw_networkx(grafo, pos, with_labels=True, node_size=1000, edge_color='gray', font_size=10, font_color='black', node_color='lightblue')
    plt.title('Grafo')

    # Estilização com Seaborn
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(False)
    plt.gca().spines['left'].set_visible(False)
    plt.xticks([])
    plt.yticks([])
    sns.despine(left=True, bottom=True)

    # Ajuste da paleta de cores
    sns.set_palette(["lightblue"])

    plt.tight_layout()
    plt.show()

    # Plotar nós mais importantes em paralelo
    top_nodes = sorted(grafo.degree, key=lambda x: x[1], reverse=True)[:top_n]
    plt.figure(figsize=(12, 8))
    pos = nx.spring_layout(grafo, seed=42, k=0.5)
    nx.draw_networkx(grafo, pos, nodelist=[node[0] for node in top_nodes], with_labels=True, node_size=1000, edge_color='gray', font_size=10, font_color='black', node_color='lightblue')
    plt.title('Nós Mais Importantes')

    # Estilização com Seaborn
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(False)
    plt.gca().spines['left'].set_visible(False)
    plt.xticks([])
    plt.yticks([])
    sns.despine(left=True, bottom=True)

    # Ajuste da paleta de cores
    sns.set_palette(["lightblue"])

    plt.tight_layout()
    plt.show()

# Função para remover stopwords, pontuação e acentuação
def preprocess_text(text):
    text = unicodedata.normalize('NFKD', text).encode('ASCII', 'ignore').decode('utf-8')
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    tokens = nltk.word_tokenize(text)
    stop_words = set(stopwords.words('portuguese'))
    tokens = [token for token in tokens if token.lower() not in stop_words]
    
    return ' '.join(tokens)

# Preprocessamento das mensagens no dataframe
df_atualizado['mensagem_preprocessada'] = df_atualizado['mensagem'].apply(preprocess_text)

# Dicionários para armazenar as contagens de ocorrências de cada n-grama
frequencia_bigramas = {}
frequencia_trigramas = {}
frequencia_quadrigramas = {}

# Loop para processar cada mensagem no dataframe
for mensagem in df_atualizado['mensagem_preprocessada']:
    # Calcular bigramas
    bigramas = calcular_ngrams(mensagem, 2)
    for bigrama in bigramas:
        if bigrama in frequencia_bigramas:
            frequencia_bigramas[bigrama] += 1
        else:
            frequencia_bigramas[bigrama] = 1

    # Calcular trigramas
   
    trigramas = calcular_ngrams(mensagem, 3)
    for trigrama in trigramas:
        if trigrama in frequencia_trigramas:
            frequencia_trigramas[trigrama] += 1
        else:
            frequencia_trigramas[trigrama] = 1

    # Calcular quadrigramas
    quadrigramas = calcular_ngrams(mensagem, 4)
    for quadrigrama in quadrigramas:
        if quadrigrama in frequencia_quadrigramas:
            frequencia_quadrigramas[quadrigrama] += 1
        else:
            frequencia_quadrigramas[quadrigrama] = 1

# Criação do grafo dos bigramas
grafo_bigramas = criar_grafo_ngramas(frequencia_bigramas.keys())
# Plotagem do grafo dos bigramas e dos nós mais importantes em paralelo
plotar_grafo(grafo_bigramas, top_n=10)

# Criação do grafo dos trigramas
grafo_trigramas = criar_grafo_ngramas(frequencia_trigramas.keys())
# Plotagem do grafo dos trigramas e dos nós mais importantes em paralelo
plotar_grafo(grafo_trigramas, top_n=10)

# Criação do grafo dos quadrigramas
grafo_quadrigramas = criar_grafo_ngramas(frequencia_quadrigramas.keys())
# Plotagem do grafo dos quadrigramas e dos nós mais importantes em paralelo
plotar_grafo(grafo_quadrigramas, top_n=10)
```

```python
from textblob import TextBlob

# Função para realizar a análise de sentimento em uma mensagem
def analisar_sentimento(mensagem):
    blob = TextBlob(mensagem)
    polaridade = blob.sentiment.polarity
    if polaridade > 0:
        return 'Positivo'
    elif polaridade < 0:
        return 'Negativo'
    else:
        return 'Neutro'

# Adicionar uma coluna 'sentimento' ao dataframe com o resultado da análise de sentimento
df_atualizado['sentimento'] = df_atualizado['mensagem'].apply(analisar_sentimento)

import seaborn as sns
import matplotlib.pyplot as plt

# Contagem dos valores únicos da coluna 'sentimento'
contagem_sentimentos = df_atualizado['sentimento'].value_counts()

# Configurar estilo do seaborn
sns.set(style="whitegrid")

# Plotar o gráfico de contagem de sentimentos
plt.figure(figsize=(8, 6))
sns.countplot(data=df_atualizado, x='sentimento')
plt.title('Distribuição de Sentimentos')
plt.xlabel('Sentimento')
plt.ylabel('Contagem')

# Mostrar o gráfico
plt.show()

```
